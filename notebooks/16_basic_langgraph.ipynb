```python
from pydantic import BaseModel
from langgraph.graph import StateGraph , START , END ### To maintain and track the record 
```

```python
class GreetState(BaseModel):
    message: str = ""
    message_1:str = ""
```

```python
graph = StateGraph(GreetState)
```

#### NODES - Python Functions 

```python
def Greet(state:GreetState):
    state.message = "Hello , How are you ??"
    state.message_1 = f"{state.message_1} ! All Good ! "
    return state
```

```python

graph.add_node("greetings_new",Greet)
```

```python
graph.add_edge(START, "greetings_new")
graph.add_edge("greetings_new",END)
```

```python
finalGraph = graph.compile()
```

```python
res = finalGraph.invoke({"message_1":"I love Kashmiri Food !"})
```
