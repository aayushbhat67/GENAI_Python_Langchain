```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool 
```

```python
@tool
def add_numbers(a:int, b:int):
    """
        it will return the sum of two numbers

        Args: 
            a : Number One 
            b : Number two
    """
    return a+b


@tool
def multiply_numbers(a:int, b:int):
    """
        it will return the product of two numbers

        Args: 
            a : Number One 
            b : Number two
    """
    return a*b
```

```python
add_numbers.invoke({"a":12,"b":3})
```

```python
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent 
```

```python
llm = ChatOpenAI(model="gpt-4")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_numbers,multiply_numbers],
    system_prompt="You are a maths teacher, and always use tool for calculation."
    
)
```

```python
response = agent.invoke({"messages":[{"role":"user","content":"What is 2 + 3 ?"}]})
response_1 = agent.invoke({"messages":[{"role":"user","content":"What is 6 * 7 ?"}]})
```

```python
for res in response['messages']:
    print(res)
    print("\n")
    
for res in response_1['messages']:
    print(res)
    print("\n")
    
```

```python
for res in response['messages']:
    print(res.content)

    
for res in response_1['messages']:
    print(res.content)
    
    
```

```python
response = agent.invoke({"messages":[{"role":"user","content":"What is 2 + 3  + 2 * 6?"}]})
response_1 = agent.invoke({"messages":[{"role":"user","content":"What is 6 * 7 + 2 * 7?"}]})

print(response["messages"][-1].content)
print(response_1["messages"][-1].content)
```
