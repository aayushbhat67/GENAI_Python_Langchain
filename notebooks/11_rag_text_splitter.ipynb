#### SPLITTER IN DOC LOADER 

```python
from langchain_community.document_loaders import TextLoader , PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("../Data/My Resume.txt")
text = loader.load()
```

```python
pdfloader = PyPDFLoader("../Data/0121YG008021121_190858c.pdf")
pdf = pdfloader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 500)
```

```python
text_chunks = splitter.split_documents(text)
len(text_chunks)
```

```python
text_chunks
```

```python
pdf_chunks = splitter.split_documents(pdf)
len(pdf_chunks)
```
