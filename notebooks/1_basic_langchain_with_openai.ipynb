#### PROMPT CHAINS - DYNAMIC PROMPTS

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
llm = ChatOpenAI(model="gpt-4o")
out = StrOutputParser()
```

#### MY SELF CHAIN MADE

```python
def transform_case(result:str):
    return result.upper()
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role":"system","content":"You are a translator and translate input into {language}."},
    {"role":"user","content":"{query}"},
])
```

```python

# res = llm.invoke(prompts.invoke({"language":"Hindi","query":"I love you ?"}))
# res.content
```

```python
chains = prompts | llm | out | transform_case
chains
```

```python
res = chains.invoke({"language":"Hinglish","query":"I love programming."})
print(res)
```
